### Задание 1.1

```python
import matplotlib
from matplotlib import pyplot as plt
matplotlib.pyplot.style.use('ggplot')
```

```python
x = range(-10,11,1)
y = [i**3 for i in x]
plt.plot(x,y,'r')
plt.show()
```

### Задание 1.2

```python
plt.plot(x,y,'r')
plt.title(u'График')
plt.xlabel(u'ось Х')
plt.ylabel(u'ось У', rotation='horizontal')
plt.show()
```

### Задание 1.3

```python
plt.plot(x,y,'r',label='$y=x^3$')
plt.title(u'График')
plt.xlabel(u'ось Х')
plt.ylabel(u'ось У', rotation='horizontal')
plt.legend(loc = 2)
plt.show()
```

### Задание 1.4

```python
plt.plot(x,y,'r',label='$y=x^3$')
plt.title(u'График')
plt.xlabel(u'ось Х')
plt.ylabel(u'ось У', rotation='horizontal')
plt.annotate(u'точка (0,0)', xy=(0, 0), xytext=(1, -250),
            arrowprops=dict(facecolor='g', shrink=0))
plt.legend(loc = 2)
plt.show()
```

### Задание 2.1

```python
x=range(-10,11,1)
y1 = [i**3 for i in x]
y2 = [2*i**3 for i in x]
```

```python
plt.plot(x,y1,'y',label='$y=x^3$')
plt.plot(x,y2,'black',label='$y=2*x^3$')
plt.fill_between(x, y1, y2, color='g', alpha=0.5, label=u'область')
plt.title(u'Графики')
plt.xlabel(u'ось Х')
plt.ylabel(u'ось У',rotation='horizontal')
plt.legend(loc = 2)
plt.show()
```

### Задание 2.2

```python


x=range(-10,11,1)
y11 = [i**2 for i in x]
y12 = [2*i**2 for i in x]
y21 = [i**3 for i in x]
y22 = [2*i**3 for i in x]

plt.figure(figsize=(12,5))

plt.subplot(121)
plt.plot(x,y11,'y',label='$y=x^2$')
plt.plot(x,y12,'black',label='$y=2*x^2$')
plt.fill_between(x, y11, y12, color='g', alpha=0.5, label=u'область')
plt.title(u'Графики 1')
plt.xlabel(u'ось Х')
plt.ylabel(u'ось У', rotation='horizontal')
plt.legend(loc = 9)

plt.subplot(122)
plt.plot(x,y21,'y',label='$y=x^3$')
plt.plot(x,y22,'black',label='$y=2*x^3$')
plt.fill_between(x, y21, y22, color='g', alpha=0.5, label=u'область')
plt.title(u'Графики 2')
plt.xlabel(u'ось Х')
plt.ylabel(u'ось У',rotation='horizontal')
plt.legend(loc = 9)

plt.show()
```
